````python
def get_press_ratings(allocine_url):
    result = requests.get(allocine_url).text
    soup = bs4.BeautifulSoup(result)
    ratings = {}
    for raw_rating in soup.findAll("li",{"class":"item"}):
        rating = raw_rating.div.span['title']
        press = raw_rating.div.findAll('span')[-1].text
        ratings[press]=rating
    return ratings
press_ratings = get_press_ratings("https://www.allocine.fr/film/fichefilm-288402/critiques/presse/")
````

````python

def get_top_one(title):
    title = title.replace(' ', '+')
    query = f"http://html.duckduckgo.com/html/?q={title}+www.allocine.fr+critiques+presse"
    #print(query)
    result = requests.get(query, headers={'user-agent': 'my-app/0.0.1'}).text
    soup = bs4.BeautifulSoup(result)
    for result in soup.findAll('a',{'class':"result__url"}):
        result = result.text.strip()
        if result.startswith('www.allocine.fr') and 'critiques' in result:
            if not result.endswith('presse/'):
                if result.endswith('spectateurs/'):
                    result = result[:-len('spectateurs/')] + 'presse/'
                else:
                    result += 'presse/'
            return "http://"+result
    return None
````
